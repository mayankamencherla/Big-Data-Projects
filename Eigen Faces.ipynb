```python
from sklearn.datasets import fetch_olivetti_faces 
faces = fetch_olivetti_faces()
```

```python
 # set up the figure 
fig = plt.figure(figsize=(6, 6))# figure size in inches 
fig.subplots_adjust(left=0, right=1, bottom=0, top=1, hspace=0.05, wspace=0.05)
# plot the digits: each image is 8x8 pixels 
for i in range(100): 
    ax = fig.add_subplot(10, 10, i + 1, xticks=[], yticks=[]) 
    ax.imshow(faces.images[i], cmap=plt.cm.gray)
# label the image with the target value 
    ax.text(0, 7, str(faces.target[i]))
    
```

```python
 faces.data.shape # 400 faces, each face has 64 * 64 = 4096 pixels
    
```

```python
n, h, w = faces.images.shape
face = faces.data[0,:].reshape(h, w) 
plt.imshow(face, cmap=plt.cm.gray) 
plt.show()

#Clearly this is a 1-D matrix called faces.data. 
#It contains 0 to 63 numbered faces as data points.
```

```python
import numpy as np 
import matplotlib.pyplot as plt 
from mpl_toolkits.mplot3d import Axes3D 
from sklearn.decomposition import PCA
```

```python
faces.data.shape
```

```python
from sklearn import decomposition 
pca = decomposition.PCA() 
```

```python
pca = PCA(n_components=200, whiten=False)
pca.fit(faces.data)


# Get the principal components
V = pca.components_
# Print the top 9 principal components (or "eigenfaces")
for i in range(9):
    face = V[i,:].reshape(h, w)
    plt.imshow(face, cmap=plt.cm.gray)
    plt.show()
```

```python
pca = PCA(n_components=9, whiten=False)
pca.fit(faces.data)

faces.data_pca = pca.transform(faces.data)

faces.data = pca.inverse_transform(faces.data_pca)

n, h, w = faces.images.shape
face = faces.data[0,:].reshape(h, w) 
plt.imshow(face, cmap=plt.cm.gray) 
plt.show()
```
