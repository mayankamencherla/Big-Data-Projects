```python
# -*- coding: utf-8 -*-
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from  sklearn import decomposition
import scipy

mean = np.array([11,8])
cov = np.array([(1,1),(1,3)])#data for 2D Gaussian

X = np.random.multivariate_normal(mean, cov,100) #create 2D Gaussian

plt.scatter(X[:,0],X[:,1]) # scatters plot
ax = plt.axes()
ax.set_xlim([3,18])
ax.set_ylim([0,15])
ax.set_aspect('equal')
plt.show()

a = X[:,0] - X[:,0].mean()
b = X[:,1] - X[:,1].mean()

X1 = np.vstack((a,b))
print X1.shape

from numpy import linalg

X_trans = X1.T
print X_trans.shape

A = np.dot(X1,X_trans)/100 #Normalizing the matrix
print A.shape

D,V = np.linalg.eig(np.cov(X1))
print D

V = V
print V
```
