The first step we will take is to understand the noise characteristics of the position estimate of a single GPS receiver.  From this, we will be able to develop a noise model which will be needed in determining when/whether the receiver is able to detect an earthquake.  We will train our model using data from the previous day, on which it is assumed there is no earthquake.

The code below will load in data from the previous day for a single receiver and plot the displacement (from the mean location) in meters in the east, north, and up directions

```python
import GEONET
from datetime import datetime, timedelta
from matplotlib import dates

# The time ranges between which data will be downloaded.
rxID = 145
tstart = datetime(2011,3,10,0,1,0) # datetime(year, month, day, hour, minute, second)
tstop =  datetime(2011,3,10,23,59,0)

# Query the database
rxdata = GEONET.get_rx_data(rxID, tstart, tstop)

# Extract useful information and plot it
posix_time = rxdata['UT']
lat        = rxdata['lat']
lon        = rxdata['lon']
alt        = rxdata['alt']

# Convert from POSIX time to Python's "datetime" object
time = []
for p in posix_time:
    time.append(datetime.utcfromtimestamp(p))

# Convert latitude, longitude, altitude to displacements in east, north, up
# (relative to an average of the first few samples)
lat_ref    = mean(lat[1:10])
lon_ref    = mean(lon[1:10])
alt_ref    = mean(alt[1:10])
delta_lon  = lon - lon_ref
delta_lat  = lat - lat_ref
delta_alt  = alt - alt_ref
east       = delta_lon * 111.0e3 * cos(lat_ref) # 111*cos(lat) km per degree longitude
north      = delta_lat * 111.0e3 # 111 km per degree latitude
up         = delta_alt

# Plot
plot(time, east,  'k.-', label = 'east')
ylabel('Displacement [meters]')
xlabel('UT')
legend(loc='best')
plt.gca().xaxis.set_major_formatter(dates.DateFormatter('%H'))

figure()
plot(time, north,  'k.-', label = 'north')
ylabel('Displacement [meters]')
xlabel('UT')
legend(loc='best')
plt.gca().xaxis.set_major_formatter(dates.DateFormatter('%H'))

figure()
plot(time, up,  'k.-', label = 'up')
ylabel('Displacement [meters]')
xlabel('UT')
legend(loc='best')
plt.gca().xaxis.set_major_formatter(dates.DateFormatter('%H'))
```

### Exercise 1: Data pre-processing

Often with Big Data applications, data quality is a major concern. As such, it is important to "clean" your data before your analysis. This is especially important when dealing with real-world sensors that have systematic error sources.

1) We are interested in detecting fast changes in the GPS receiver's position. As you can see with the plots above, there are slowly-varying features in the data, caused by the error sources discussed in class (the troposphere, the ionosphere, multipath, etc.) 

To remove these slowly-varying features, which are common across all the receivers, we will high-pass filter the data.  There are many different ways to filter the data, and you may develop whichever method you would like.  However, one simple method would be to create a sliding window averager in which the mean of the previous N samples are subtracted from the current sample. A good choice for N might be 20, which would remove the mean from the last 10 minutes of data.

We've provided the template for a function that will query the database and do the filtering. Please replace the section labeled ##### PLACE YOUR FILTER HERE ##### with your filter.

```python
# Function to query for a day's data and return the filtered fluctuations
def get_fluctuations(rxID, tstart, tstop):
    '''
    Function to return the displacements in the east (x), north (y), and up (z) directions for the requested receiver (rxID)
    between two times (tstart, stop).
    '''
    
    # Initialize return variables
    time = []
    x = []
    y = []
    z = []
    
    # Query the database
    rxdata = GEONET.get_rx_data(rxID, tstart, tstop)
    
    # Extract useful information
    posix_time = rxdata['UT']
    lat        = rxdata['lat']
    lon        = rxdata['lon']
    alt        = rxdata['alt']
    
    # Convert from POSIX time to Python's "datetime" object
    time = []
    for p in posix_time:
        time.append(datetime.utcfromtimestamp(p))
        
    L = len(time)
    
    # Convert latitude, longitude, altitude to displacements in east, north, up
    # (relative to an average of the first few samples)
    lat_ref    = mean(lat[1:10])
    lon_ref    = mean(lon[1:10])
    alt_ref    = mean(alt[1:10])
    delta_lon  = lon - lon_ref
    delta_lat  = lat - lat_ref
    delta_alt  = alt - alt_ref
    east       = 111.0e3 * cos(lat_ref) * delta_lon  # 111*cos(lat) km per degree longitude
    north      = 111.0e3 * delta_lat # 111 km per degree latitude
    up         = delta_alt
    
    x = np.zeros(len(east))
    y = np.zeros(len(north))
    z = np.zeros(len(up))
    ##### PLACE YOUR FILTER HERE #####
    # Filter east, north, up to get x,y,z
    N = 20
    for i in range(len(time)):
        if i ==0 or i==1:
            x[i] = east[i]
            y[i] = north[i] 
            z[i] = up[i]
        elif i <= 20 and i>1:
            x[i] = east[i] - mean(east[0:i-1])
            y[i] = north[i] - mean(north[0:i-1])
            z[i] = up[i] - mean(up[0:i-1])
        else:
            x[i] = east[i] - mean(east[i-20:i-1])
            y[i] = north[i] - mean(north[i-20:i-1])
            z[i] = up[i] - mean(up[i-20:i-1])
    
    return time,x,y,z
```

2) Derive the maximum a posteriori (MAP) decision rule for deciding if an earthquake happened at a given time. You should find that the decision rule boils down to simply comparing $|x|$ to a constant threshold, where $x$ is the eastward perturbation. The hypotheses are:

$H_0$: No earthquake is present. As you found in Exercise 1, part 2, from the training data on March 10, $x$ is a random variable, distributed as a Gaussian with mean 0 and some standard deviation, $\sigma$, which you found for one particular receiver.

$H_1$: An earthquake has occurred. Let's assume that the earthquake could have any amplitude between -10 meters and +10 meters, with uniform probability.

Recall that the MAP decision rule is to choose $H_1$ if and only if:

$$ \frac{p(x|H_1)}{p(x|H_0)} > \frac{\pi_0}{\pi_1} $$

where $\pi_1$ is the prior probability that an earthquake occurs at a given time.
