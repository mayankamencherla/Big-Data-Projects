## B. Working with the GEONET GPS data

We have stored two days' worth of GEONET GPS data in an online database housed on one of our servers.  The data are comprised of estimates of a receiver's latitude, longitude, and altitude as a function of time.  The temporal resolution for the data from each receiver is 30 seconds.  The data are processed using a differential technique, meaning that one receiver within the network is chosen as a "reference" and all of the other navigation solutions are calculated based from that reference station.  The advantage of this sort of technique is that all common sources of error and bias (e.g., satellite clock errors, orbital errors) in the GPS solution are removed.  However, errors that are not shared between the receivers (e.g., local multipath environment, local positional changes) are propagated from the reference receiver into the rest of the network.

The database can be queried using SQL commands.  Alternately, you can use the functions provided in the GEONET.py module, available in the zip file that came with the lab. It is suggested that you use these functions provided to you.  However, you should read through the code to make sure you understand how the data are being accessed and stored.  Below, we give you examples of using the GEONET.py module to perform basic tasks. Make sure that the GEONET module is either in the same directory that this notebook is running in, or on the Python path.

```python
# Extract useful information and plot it
posix_time = rxdata['UT']
lat        = rxdata['lat']
lon        = rxdata['lon']
alt        = rxdata['alt']
```

Now, we can plot the displacement (from some mean location), as shown below.  This displacement is either related to the physical displacement of the GPS antenna or noise in the system.

```python
# Convert from POSIX time to Python's "datetime" object
t = []
for p in posix_time:
    t.append(datetime.datetime.utcfromtimestamp(p))

# Convert latitude, longitude, altitude to displacements in east, north, up
# (relative to an average of the first few samples)
lat_ref    = mean(lat[1:10])
lon_ref    = mean(lon[1:10])
alt_ref    = mean(alt[1:10])
delta_lon  = lon - lon_ref
delta_lat  = lat - lat_ref
delta_alt  = alt - alt_ref
east       = delta_lon * 111.0e3 * cos(lat_ref) # 111*cos(lat) km per degree longitude
north      = delta_lat * 111.0e3 # 111 km per degree latitude
up         = delta_alt

# Plot
subplot(311)
plot(t, east,  '.-', label = 'east')
legend(loc='best')
subplot(312)
plot(t, north, '.-', label = 'north')
legend(loc='best')
subplot(313)
plot(t, up   , '.-', label = 'up')
ylabel('Displacement [meters]')
xlabel('UT')
legend(loc='best')
plt.gca().xaxis.set_major_formatter(dates.DateFormatter('%H:%M'))
```

For this section, we suggest working with receivers [172, 901, 1162, 205].  Your model can be a simple one, and we would suggest:

$ time of arrival = time of earthquake + distance/velocity 

$ t_1 = t_e + v^{-1} \sqrt{(x_1 - x_e)^2 + (y_1 - y_e)^2} $

and 3 more equations, with $t_2, t_3, t_4, x_2, x_3, x_4, y_2, y_3, y_4$, all known. The unknowns are $t_e, x_e, y_e$, and $v$.

One way to solve these equations is with an iterative approach, identical to the one described in lecture to solve for a GPS position and clock offset. Another way to solve this problem is with a more brute force approach. Either is acceptable, but we'll describe the brute force approach here.

We'll assume the epicenter is known, and then solve for $t_e$ and $v$. We'll try a bunch of trial values of the epicenter, and see which ones agrees with our data the best.

Assuming that $x_e$ and $y_e$ are known, then let's define:

$d_i = \sqrt{(x_i - x_e)^2 + (y_i - y_e)^2}$

where $i$ is 1,2,3, or 4. Note that $d_i$ is known. The equation above then reduces to:

$$
t_i = t_e + v^{-1} d_i ~~~~~~~ i=1,2,3,4
$$

These four equations can be written in matrix notation as:

$$
\begin{bmatrix}
t_1 \\
t_2 \\
t_3 \\
t_4 \\
\end{bmatrix}
=
\begin{bmatrix}
1 & d_1 \\
1 & d_2 \\
1 & d_3 \\
1 & d_4 \\
\end{bmatrix}
\begin{bmatrix}
t_e \\
v^{-1} \\
\end{bmatrix}
$$

We can estimate $t_e$ and $v^{-1}$ using the least squares solution to this matrix equation. The residual of this equation gives us an indication of the cost of the fit. A good solution will be when the cost is low.

Repeat this process for many different choices of the epicenter location, and see which has the lowest cost

```python
from numpy import linalg

t = np.array([347,346.4,347,348])
loc = np.array([40.5,38.9,38.5,37.3])

#to make an assumption of the epicenter - I'm assuming that the epicenter is at latitudes between 38 and 39
epi = linspace(35,40,51)

#Let unknown matrix dimensions 2by1 be X
cost =0
ori = 0
time = 0

#I'm going to write a for loop to loop through the 10 estimated latitudes and reduce cost.
for i in range(51):
    d0 = np.absolute(loc - epi[i]) #in meters
    d = np.array([(1,d0[0]),(1,d0[1]),(1,d0[2]),(1,d0[3])])
    
    #distance in units of degrees and time in minutes, we expect v to be of units degrees/minute and time to be in minutes
    
    X = np.linalg.lstsq(d,t)
    #From this we have X[0][0] = t and X[1] = 1/V. To compute cost:
    diff = np.absolute(t - X[0][0])
    cost_temp = np.linalg.norm(diff)
    
    if i ==0:
        cost = cost_temp
    else:   
        if(cost > cost_temp):
            cost = cost_temp
            ori = epi[i]
            time = X[0][0]
    
print "The value of latitude that minimizes cost is =%.2f" %ori
print "The value of cost for that epicenter is =%.2f" %cost
print "The time of the original quake is =%.3f" %X[0][0], "minutes"
print "Estimated velocity of seismic wave is %.2f degrees/minute" %X[0][1]
    
```

```python
# Parameters for the map
lat =    [132, 140]   # [min lat, max lat] for the map
lon =    [30, 50]   # [min lon, max lon] for the map
```

```python
# Specify the time to map.  Remember that data are only available every 30 seconds (:00 and :30).  
# You can use the same time tbefore and tafter to plot a single time "slice".
tbefore = datetime.datetime(2011,3,11,5,45,0)
tafter  = datetime.datetime(2011,3,11,5,46,0)

data_before = GEONET.get_geonet_data_at_time(tbefore)
data_after  = GEONET.get_geonet_data_at_time(tafter)

rxid = []
lat =  []
lon =  []
dlat = []
dlon = []

count = 0
rxid1vec = data_before['rxID']
rxid2vec = data_after ['rxID']
# Search through the lists and come up with a horizontal displacement for each receiver
for i in range(len(data_before)):
    rxid1 = rxid1vec[i]
    # Find the entry in data_after, if it exists:
    matches = find(rxid2vec == rxid1)
    if matches:
        j = matches[0]
        lat1 = data_before['lat'][i]
        lat2 = data_after ['lat'][j]
        lon1 = data_before['lon'][i]
        lon2 = data_after ['lon'][j]
        lat.append(lat1)
        lon.append(lon1)
        dlat.append(lat2-lat1)
        dlon.append(lon2-lon1)
        rxid.append(rxid1)
            
lat = array(lat)
lon = array(lon)
dlat = array(dlat)
dlon = array(dlon)
            
# Convert dlat, dlon to deast, dnorth
deast =  dlon * 111.0e3 * cos(lat)
dnorth = dlat * 111.0e3
```

```python
# import the library that will draw the Japanese coastline
# See http://matplotlib.org/basemap/api/basemap_api.html
from mpl_toolkits.basemap import Basemap

# Settings for plot
figsize=(8,9.7)
sc = figsize[1]/(10*5) # 1/10th of the plot width <==> 5 m, for quiver
pivot = 'middle' # for quiver
headwidth = 3
headlength = 3
headaxislength= headlength-1
minshaft = 2
qcolor = 'k'
linewidth = 0.5
qklat = 39
qklon = 136


# initialize the figure
fig = figure(num=None, figsize=figsize, dpi=200, facecolor='w', edgecolor='r')
# define the size of the region of Earth to draw, and other options:
m = Basemap(llcrnrlon=135,llcrnrlat=33,
            urcrnrlon=143,urcrnrlat=41,
            projection='merc', area_thresh=1000, resolution='i',
            lat_1=45.,lat_2=55,lat_0=40,lon_0=-85.)

for la, lo, de, dn in zip(lat, lon, deast, dnorth):
    x,y = m(lo,la)
    dx = sc*de
    dy = sc*dn
    Q = m.quiver(x, y, dx, dy, angles='uv', scale_units='inches', scale=1, 
                  pivot=pivot, headwidth=headwidth, headlength=headlength,
                  minshaft=minshaft, headaxislength=headaxislength, color=qcolor,
                  linewidth=linewidth)
xq,yq = m(qklon,qklat)
qk1 = plt.quiverkey(Q, xq, yq, sc*5, '5 m', color='k', coordinates='data')
# draw coastlines and fill the continents (alpha value is set to partial transparancy because for some
# reason the fill is over top the quivers used to denote the wind vectors.
m.drawcoastlines()

# draw parallels and meridians.
# label parallels on left and bottom
# meridians on bottom and left
# labels = [left,right,top,bottom]
parallels = np.arange(0.,81,5.)
m.drawparallels(parallels,labels=[True,False,False,True]);
meridians = np.arange(0.,355.,5.)
m.drawmeridians(meridians,labels=[True,False,False,True]);

title('Earthquake Displacement')
```

```python
#PART A:
tbefore = datetime.datetime(2011,3,10,3,0,0)
tafter  = datetime.datetime(2011,3,10,9,0,0)

data_before = GEONET.get_geonet_data_at_time(tbefore)
data_after  = GEONET.get_geonet_data_at_time(tafter)

rxid = []
lat =  []
lon =  []
dlat = []
dlon = []

count = 0
rxid1vec = data_before['rxID']
rxid2vec = data_after ['rxID']
# Search through the lists and come up with a horizontal displacement for each receiver
for i in range(len(data_before)):
    rxid1 = rxid1vec[i]
    # Find the entry in data_after, if it exists:
    matches = find(rxid2vec == rxid1)
    if matches:
        j = matches[0]
        lat1 = data_before['lat'][i]
        lat2 = data_after ['lat'][j]
        lon1 = data_before['lon'][i]
        lon2 = data_after ['lon'][j]
        lat.append(lat1)
        lon.append(lon1)
        dlat.append(lat2-lat1)
        dlon.append(lon2-lon1)
        rxid.append(rxid1)
                
lat = array(lat)
lon = array(lon)
dlat = array(dlat)
dlon = array(dlon)
                
# Convert dlat, dlon to deast, dnorth
deast =  dlon * 111.0e3 * cos(lat)
dnorth = dlat * 111.0e3
    # Settings for plot
figsize=(8,9.7)
sc = figsize[1]/(10*5) # 1/10th of the plot width <==> 5 m, for quiver
pivot = 'middle' # for quiver
headwidth = 3
headlength = 3
headaxislength= headlength-1
minshaft = 2
qcolor = 'k'
linewidth = 0.5
qklat = 39
qklon = 136
    

# initialize the figure
fig = figure(num=None, figsize=figsize, dpi=200, facecolor='w', edgecolor='r')
# define the size of the region of Earth to draw, and other options:
m = Basemap(llcrnrlon=135,llcrnrlat=33,
            urcrnrlon=143,urcrnrlat=41,
            projection='merc', area_thresh=1000, resolution='i',
            lat_1=45.,lat_2=55,lat_0=40,lon_0=-85.)

for la, lo, de, dn in zip(lat, lon, deast, dnorth):
    x,y = m(lo,la)
    dx = sc*de
    dy = sc*dn
    Q = m.quiver(x, y, dx, dy, angles='uv', scale_units='inches', scale=1, 
                  pivot=pivot, headwidth=headwidth, headlength=headlength,
                  minshaft=minshaft, headaxislength=headaxislength, color=qcolor,
                  linewidth=linewidth)
xq,yq = m(qklon,qklat)
qk1 = plt.quiverkey(Q, xq, yq, sc*5, '5 m', color='k', coordinates='data')
# draw coastlines and fill the continents (alpha value is set to partial transparancy because for some
# reason the fill is over top the quivers used to denote the wind vectors.
m.drawcoastlines()

# draw parallels and meridians.
# label parallels on left and bottom
# meridians on bottom and left
# labels = [left,right,top,bottom]
parallels = np.arange(0.,81,5.)
m.drawparallels(parallels,labels=[True,False,False,True]);
meridians = np.arange(0.,355.,5.)
m.drawmeridians(meridians,labels=[True,False,False,True]);

title('Earthquake Displacement')
```
